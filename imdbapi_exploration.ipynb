```python
def get_movies(person):
    """Returns a list of tuples of movie_ID's and titles from a person object"""
    movie_ids = [film.movieID for film in get_actor_movie(person)]
    movie_titles = [film['title'] for film in get_actor_movie(person)]
    movies_list = [i for i in zip(movie_ids, movie_titles)]
    return movies_list
```

I need to split the strings at the delimiter and put the corresponding values in a list of tuples to make DataFrame columns.

```python
def salary_tuples(person):
    """Returns a list of tuples of projects and salaries from person object"""
    # check to see if salary data exists
    if person.get('salary'):
        # split data at demarcation
        clean_lists = [item.split('::') for item in person.get('salary')]
        # sort into two lists, one for titles and one for salaries
        split_salary_lists = [[ i for i, j in clean_lists ], [ j for i, j in clean_lists ]]
        # make list of tuples with corresponding data pairs
        tuples_list = list(zip(split_salary_lists[0], split_salary_lists[1]))
        return tuples_list
```

```python
# observe pitt salary tuples
salary_tuples(pitt)
```

```python
# observe reeves salary tuples
salary_tuples(reeves)
```

Unfortunately, the delimiters of the salary strings are different between pitt object and reeves object.<br>
For now, I will split the string on these two delimiters by calling for their intersection. In the future, I should have a way to do it for all potential cases.

Now, unpack them into a DataFrame while cleaning them up.

```python
def salaries_df(person):
    """Returns a DataFrame with clean movie titles and clean salary as a float for person object with salary data"""
    # check if person object has salary data
    if person.get('salary'):
        # unpack salary tuples into DataFrame columns
        df = pd.DataFrame(salary_tuples(person), columns=['Title', 'Salary'])
        # strip salary of monetary punctuation
        df['Salary'] = df['Salary'].str.replace(',|\$', '')
        # split off salary comments and return dollar amount as a float
        df['Salary'] = df['Salary'].str.split('[(|+]', expand=True)[0].astype(float)
        # remove release dates
        df['Title'] = df['Title'].str.split('\s+\(|\s\+').str[0]
        return df
```

Now, put this all together into their DataFrames

```python
def merge_salaries_df(person):
    """Merges salaries DatFrame into person DataFrame on Title"""
    # chek if person has salary data
    if person.get('salary'):
        # perform outer merge on the two DataFrames to retain all the movieID's
        df = pd.merge(movies_df(person), salaries_df(person), on='Title', how='outer')
        return df
    else:
        # return the DataFrame with no salary data as unchanged
        return movies_df(person)
```
